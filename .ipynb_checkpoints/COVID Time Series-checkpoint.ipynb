## 1. Analyze countries like China, South Korea, USA, Italy and Spain individually to see increasing number of cases on day-to-day basis. Provide insight and feedback using a deck as if you were providing findings to Senior management

```python
countries = ['Mainland China', 'US', 'Italy', 'Spain', 'Korea, South']


five_countries = covid[covid['Country/Region'].isin(countries)]
```

```python
grouped = five_countries.groupby(['Country/Region','ObservationDate']).sum()
```

```python
grouped.drop(columns=['SNo'], inplace=True)
```

```python
for country in countries:
    sns.lineplot(data = grouped, x = 'ObservationDate', y = 'Confirmed', hue ='Country/Region')
    plt.title('Total COVID Cases')
    plt.ylabel('Confirmed Cases')
    plt.xlabel('Date')
    plt.legend(countries)
```

```python
for country in countries:
    daily_change = [0]
    for date in range(1 ,len(grouped['ObservationDate'])):
        daily_change.append((grouped[country][date] - grouped[country][date-1]))
    grouped[f'{country}_daily_change'] = daily_change
```

```python
change_cols = ['China_daily_change', 'US_daily_change','Italy_daily_change','Spain_daily_change', 'Korea, South_daily_change']
```

```python
for country in change_cols:
    sns.lineplot(data=five_countries, x = 'Date', y = country)
    plt.title('Daily Change in Number of COVID Cases')
    plt.ylabel('Change')
    plt.xlabel('Date')
    plt.legend(countries)
```
